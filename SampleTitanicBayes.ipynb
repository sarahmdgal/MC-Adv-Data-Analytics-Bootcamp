Gaussian Naive Bayes Test

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
titanic_data_2.count()
```

```python
titanic_data_2.dtypes
```

```python
titanic_data_2.corr()
```

```python
X = titanic_data_2.drop('survived', axis=1)
y = titanic_data_2['survived']
```

```python
titanic_data_2.survived.unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=50)
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train, y_train)
```

```python
gnb.score(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
cm = pd.DataFrame(
    confusion_matrix(y_test, y_pred),
    columns=['Predicted Failed', 'Predicted Passed'],
    index=['True Failed', 'True Passed']
)
```

```python
y_test.value_counts()
```

```python
gnb.score(X_test, y_test)
```

```python
print(classification_report(y_test, y_pred))
```
