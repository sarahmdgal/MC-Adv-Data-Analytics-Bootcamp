```python
lsdateStop = []
lstimeStop =[]
lsagency = []
lssubagency = []
lsdescription = []
lslocation = []
lslatitude = []
lslongitude = []
lsvehicletype = []
lsvehicleyear = []
lsvehiclemake = []
lsvehiclemodel = []
lsvehiclecolor = []
lsviolationtype = []
lsrace = []
lsgender = []
lsdriverstate = []
lsdlstate = []
lsarresttype = []
```

```python
for violation in traffic_violations:
    lsdateStop.append(violation[8])
    lstimeStop.append(violation[9])
    lsagency.append(violation[10])
    lssubagency.append(violation[11])
    lsdescription.append(violation[12])
    lslocation.append(violation[13])
    lslatitude.append(violation[14])
    lslongitude.append(violation[15])
    lsvehicletype.append(violation[27])
    lsvehicleyear.append(violation[28])
    lsvehiclemake.append(violation[29])
    lsvehiclemodel.append(violation[30])
    lsvehiclecolor.append(violation[31])
    lsviolationtype.append(violation[32])
    lsrace.append(violation[36])
    lsgender.append(violation[37])
    lsdriverstate.append(violation[39])
    lsdlstate.append(violation[40])
    lsarresttype.append(violation[41])
```

```python
lstViolations = list(zip(lsdateStop, lstimeStop, lsagency, lssubagency,
                        lsdescription, lslocation, lslatitude,lslongitude,
                        lsvehicletype, lsvehicleyear, lsvehiclemake, lsvehiclemodel,
                        lsvehiclecolor, lsviolationtype, lsrace, lsgender,
                        lsdriverstate, lsdlstate, lsarresttype))
```

```python
colnames = ['StopDate', 'StopTime', 'Agency', 'SubAgency', 'Description',
            'Location', 'Latitude', 'Longitude', 'VehicleType',
            'VehicleYear', 'VehicleMake', 'VehicleModel', 'VehicleColor',
            'ViolationType', 'EthnicityRace', 'Gender', 'DriverState',
            'DrvLicState','ArrestType']
```

```python
df = pd.DataFrame(lstViolations, columns=colnames)

df.head(10)
```
