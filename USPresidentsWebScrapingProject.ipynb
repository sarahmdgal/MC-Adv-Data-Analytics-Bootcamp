```python
import urllib.request as urllib2
from urllib.request import Request, urlopen
from bs4 import *
import pandas as pd
```

```python
req = Request("https://www.loc.gov/rr/print/list/057_chron.html", headers={'User-Agent': 'Mozilla/65.0.1'})
webpage = urlopen(req).read()


 
```

```python
print(webpage)
```

```python
soup = BeautifulSoup(webpage, "html.parser")
```

```python
all_links=soup.find_all("a")
for link in all_links:
    print(link.get('href'))
```

```python
all_tables=soup.find_all('table')
all_tables
```

```python
right_table=soup.find('table')
right_table
```

```python
lstYears=[]
lstPresidentName=[]
lstPolitPartyAffil=[]
lstVicePresident=[]
```

```python
for row in right_table('tr'):
    cells = row.find_all('td')
    if len(cells)==4:
        lstYears.append(cells[0].find(text=True))
        lstPresidentName.append(cells[1].find(text=True))
        lstPolitPartyAffil.append(cells[2].find(text=True))
        lstVicePresident.append(cells[3].find(text=True))
```

```python
lstYears
```

```python
lstPresidentsInfo = list(zip(lstYears, lstPresidentName, lstPolitPartyAffil, lstVicePresident))
```
