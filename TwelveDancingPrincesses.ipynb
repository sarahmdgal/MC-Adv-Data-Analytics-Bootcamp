```python
import pandas as pd
import sys
import numpy as np
import scipy as sc
import bs4 as soup
import nltk
import codecs
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.tokenize import TweetTokenizer
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer

from nltk.corpus import names
from string import punctuation

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('names')

```

```python
fin = open('datasets_12dancingprincesses.txt')
princesses = fin.read()
```

```python
word_tokenize(princesses)
```

```python
sepwords = princesses.split()
```

```python
sepwords
```

```python
princesses_word_count = FreqDist(sepwords)
```

```python
princesses_word_count
```

```python
princesses_word_count.most_common(10)
```

```python
len(princesses_word_count)
```

```python
for word in princesses_word_count:
    if word in punctuation:
        princesses_word_count.remove(word)
```

```python
len(princesses_word_count)
```

```python
punctuation
```

```python
eng_stopwords = stopwords.words('english')
eng_stopwords
```

```python
rm_count = 0
new_words = []

for word in princesses_word_count:
    if word not in eng_stopwords:
        new_words.append(word)
    else:
        rm_count += 1
```
